```python
import json
import pandas as pd
import datetime

class GooglePlaces(object):
    def __init__(self, apiKey):
        self.apiKey = apiKey

    def search_places_by_coordinate(self, location, radius, types):
        """
        Search for places based on geographic coordinates.

        Args:
            location (str): The geographic coordinates (latitude, longitude) around which to search.
            radius (str): The radius (in meters) around the specified location to search within.
            types (str): The type of places to search for (e.g., "restaurant").

        Returns:
            list: A list of places found within the specified radius of the given location.
        """
        endpoint_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
        places = []
        params = {
            'location': location,
            'radius': radius,
            'type': types,
            'key': self.apiKey
        }
        res = requests.get(endpoint_url, params=params)
        results = json.loads(res.content)
        places.extend(results['results'][:3])  # Limit to first 3 places
        return places

    def get_place_details(self, place_id, fields):
        """
        Retrieve details for a specific place.

        Args:
            place_id (str): The unique identifier for the place.
            fields (list): A list of fields to retrieve for the place.

        Returns:
            dict: Details for the specified place.
        """
        endpoint_url = "https://maps.googleapis.com/maps/api/place/details/json"
        params = {
            'place_id': place_id,
            'fields': ",".join(fields),
            'key': self.apiKey
        }
        res = requests.get(endpoint_url, params=params)
        place_details = json.loads(res.content)
        return place_details

# Inicializa GooglePlaces con una clave API
api = GooglePlaces("AIzaSyC7nVxA_-00iMv7fnNkErI8zSeshb7C4ek")

# Busca restaurantes cercanos
places = api.search_places_by_coordinate("-12.084741735171747, -77.0355305606113", "500", "restaurant")
fields = ['name', 'formatted_address', 'international_phone_number', 'website', 'rating', 'review', 'opening_hours']

restaurants_info = []

# Itera sobre los restaurantes encontrados
for place in places:
    # Obtiene los detalles del restaurante
    details = api.get_place_details(place['place_id'], fields)
    # Extrae la información del restaurante
    restaurant_details = {
        'name': details['result'].get('name', ''),
        'address': details['result'].get('formatted_address', ''),
        'phone_number': details['result'].get('international_phone_number', ''),
        'website': details['result'].get('website', ''),
        'rating': details['result'].get('rating', ''),
        'monday_opening_hours': '',
        'tuesday_opening_hours': '',
        'wednesday_opening_hours': '',
        'thursday_opening_hours': '',
        'friday_opening_hours': '',
        'saturday_opening_hours': '',
        'sunday_opening_hours': '',
    }
    # Si hay información sobre las horas de apertura, la formatea y la agrega a los detalles del restaurante
    if 'opening_hours' in details['result']:
        opening_hours = details['result']['opening_hours']['periods']
        for period in opening_hours:
            open_time = datetime.datetime.strptime(period['open']['time'], "%H%M").strftime("%I:%M %p")
            close_time = datetime.datetime.strptime(period['close']['time'], "%H%M").strftime("%I:%M %p")
            day = period['open']['day']
            weekday = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday'][day]
            restaurant_details[f'{weekday}_opening_hours'] = f"{open_time} - {close_time}"

    # Agrega los detalles del restaurante a la lista
    restaurants_info.append(restaurant_details)
    
    

# Define el directorio de salida
output_dir = r'../../output'

# Parte 1: Guardar la información en un archivo JSON
json_file_path = f'{output_dir}/restaurants_info.json'
with open(json_file_path, 'w', encoding='utf-8') as f:
    json.dump(restaurants_info, f, indent=4, ensure_ascii=False)  # Asegura que los caracteres no ASCII se guarden correctamente

# Parte 2: Crear un archivo CSV a partir de la información JSON
csv_file_path = f'{output_dir}/restaurants_info.csv'
df = pd.DataFrame(restaurants_info)
df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')
```

```python
import json
import pandas as pd
import datetime

class GooglePlaces(object):
    def __init__(self, apiKey):
        self.apiKey = apiKey

    def search_places_by_coordinate(self, location, radius, types):
        endpoint_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
        places = []
        params = {
            'location': location,
            'radius': radius,
            'type': types,
            'key': self.apiKey
        }
        res = requests.get(endpoint_url, params=params)
        results = json.loads(res.content)
        places.extend(results['results'][:3])  # Limit to first 3 places
        return places

    def get_place_details(self, place_id, fields):
        endpoint_url = "https://maps.googleapis.com/maps/api/place/details/json"
        params = {
            'place_id': place_id,
            'fields': ",".join(fields),
            'key': self.apiKey
        }
        res = requests.get(endpoint_url, params=params)
        place_details = json.loads(res.content)
        return place_details

# Initialize GooglePlaces with an API key
api = GooglePlaces("AIzaSyC7nVxA_-00iMv7fnNkErI8zSeshb7C4ek")

# Search for nearby restaurants
places = api.search_places_by_coordinate("-12.084741735171747, -77.0355305606113", "500", "restaurant")
fields = ['name', 'formatted_address', 'international_phone_number', 'website', 'rating', 'review', 'opening_hours']

restaurants_info = []

# Iterate over the found restaurants
for place in places:
    # Get details of the restaurant
    details = api.get_place_details(place['place_id'], fields)
    
    # Extract restaurant information
    restaurant_details = {
        'name': details['result'].get('name', ''),
        'address': details['result'].get('formatted_address', ''),
        'phone_number': details['result'].get('international_phone_number', ''),
        'website': details['result'].get('website', ''),
        'rating': details['result'].get('rating', ''),
        'opening_hours': '',  # Initialize opening hours
        'reviews': []  # Initialize reviews list
    }
    
    # If there are opening hours information, format and add it to the restaurant details
    if 'opening_hours' in details['result']:
        opening_hours = details['result']['opening_hours']['periods']
        opening_hours_formatted = {}
        for period in opening_hours:
            open_time = datetime.datetime.strptime(period['open']['time'], "%H%M").strftime("%I:%M %p")
            close_time = datetime.datetime.strptime(period['close']['time'], "%H%M").strftime("%I:%M %p")
            day = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'][period['open']['day']]
            opening_hours_formatted[f'{day}_opening_hours'] = f"{open_time} - {close_time}"
        restaurant_details['opening_hours'] = opening_hours_formatted
    
    # If there are reviews, add them to the restaurant details
    if 'reviews' in details['result']:
        for idx, review in enumerate(details['result']['reviews'], start=1):
            review_info = {
                f'review_{idx}_text': review['text'],
                f'review_{idx}_author': review['author_name'],
                f'review_{idx}_rating': review['rating']
            }
            restaurant_details['reviews'].append(review_info)
    
    # Add restaurant details to the list
    restaurants_info.append(restaurant_details)

# Define the output directory
output_dir = r'../../output'

# Part 1: Save information to a JSON file
json_file_path = f'{output_dir}/restaurants_info.json'
with open(json_file_path, 'w', encoding='utf-8') as f:
    json.dump(restaurants_info, f, indent=4, ensure_ascii=False)

# Part 2: Create a CSV file from JSON information
csv_file_path = f'{output_dir}/restaurants_info.csv'
df = pd.json_normalize(restaurants_info)  # Convert nested JSON to flat DataFrame
df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')
```

```python
# Function to format a review text for improved readability
def format_review(text):
    # Remove line breaks and duplicate periods
    text = text.replace('\n', ' ').replace('. .', '.')
    
    # Split the text into sentences
    sentences = text.split('. ')
    
    # Remove additional whitespace around sentences
    sentences = [sentence.strip() for sentence in sentences]
    
    # Capitalize the first letter of each sentence
    formatted_sentences = []
    for sentence in sentences:
        if sentence:
            formatted_sentence = sentence[0].upper() + sentence[1:]
            formatted_sentences.append(formatted_sentence)
    
    # Construct the formatted text
    if len(formatted_sentences) == 1:
        formatted_text = text
    else:
        formatted_text = '. '.join(formatted_sentences)
        
    # Add a period at the end if formatted_text does not end with a punctuation mark
    if formatted_text[-1] not in ['.', '!', '?']:
        formatted_text += '.'
    
    return formatted_text

# Function to determine if a restaurant point is inside the San Isidro polygon
def is_inside_san_isidro(restaurant_point, sanisidro_polygon):
    return sanisidro_polygon.contains(restaurant_point)
```

Get data from the same database used by Google Maps. Places features over 100 million businesses and points of interest that are updated frequently through owner-verified listings and user-moderated contributions.
Link: https://console.cloud.google.com/marketplace/product/google/places-backend.googleapis.com

```python
class GooglePlaces(object):
    def __init__(self, apiKey):
        self.apiKey = apiKey

    def search_places_by_coordinate(self, location, radius, types):
        endpoint_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
        places = []
        params = {
            'location': location,
            'radius': radius,
            'type': types,
            'key': self.apiKey,
            'language': 'es'  # Specify language as Spanish
        }
        res = requests.get(endpoint_url, params=params)
        results = json.loads(res.content)
        places.extend(results['results'][:3])  # Limit to first 3 places
        return places
    
    def text_search(self, query, location, types):
        endpoint_url = "https://maps.googleapis.com/maps/api/place/textsearch/json"
        places = []
        params = {
            'query': query,
            'location': location,
            'type': types,
            'key': self.apiKey,
            'language': 'es'  # Specify language as Spanish
        }
        res = requests.get(endpoint_url, params=params)
        results = json.loads(res.content)
        places.extend(results['results'][:3])  # Limit to first 3 places
        return places

    def get_place_details(self, place_id, fields):
        endpoint_url = "https://maps.googleapis.com/maps/api/place/details/json"
        params = {
            'place_id': place_id,
            'fields': ",".join(fields),
            'key': self.apiKey,
            'language': 'es'  # Specify language as Spanish
        }
        res = requests.get(endpoint_url, params=params)
        place_details = json.loads(res.content)
        return place_details

    def is_in_sanisidro(point, sanisidro_polygon):
        return sanisidro_polygon.contains(point)

```
